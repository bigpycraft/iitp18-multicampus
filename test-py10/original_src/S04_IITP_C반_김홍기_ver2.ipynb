```python
from datetime import datetime

def getMemorialDay(year, month, day, mem_day='기념임', is_msg=True):
    
    today=datetime.now()
    m_day=datetime(year, month, day)
    elapsed = today - m_day
    remain_day=abs(elapsed.days)
    remain_month=int(abs(remain_day/30))
    remain_year=int(abs(remain_month/12))
    number=list(str(remain_day))# 3자리 마다 ,찍어주도록 loop
    result=list()
    temp=''
    n=len(number)
    for i in range(1,len(number)+1):
        temp+=number.pop(-1)
        if i%3==0 and i!=n:
            temp+=','
        else:
            pass
    result=temp[::-1]          # 3자리 마다 ,찍어주도록 loop 끝
    
    if elapsed.days<0: #경과인지 남은날짜인지 판정
        if is_msg:
            print('"%s"까지 %s일 남았습니다.'%(mem_day,remain_day))
        else:
            return remain_year, remain_month, remain_day
    elif elapsed.days==0:
        if is_msg:
            print('오늘입니다')
        else:
            return remain_year, remain_month, remain_day   
            
    else:
        if is_msg:
            print('오늘은 "%s"로부터 %s일 경과되었고, 달 수로는 %d개월째, 연 수로는 %d년째 되었습니다!!'%(mem_day,result,remain_month,remain_year))
        else:
            return remain_year, remain_month, remain_day


getMemorialDay(2014,4,16,'세월호침몰사고일')
getMemorialDay(2018,12,25,'올해 크리스마스')
```

```python
from functools import *
data = ["cat", "cat", "cat", "sheep", "sheep", "duck", "duck", "duck","duck"] #입력값
data = sorted(data)                                  # reduce로 순서대로 판정하기위해 sorting 수행
temp=list(set(data))                                 # Counting용 사전형 table 생성
element=dict((i,1) for i in temp)                    # 임시로 생성된 사전형 각 key값에 value '1'로 부여하여 생성

def histoCounter(x,y):                               # counting 함수 선언
    if x==y:
        element[x]+=1
        return x
    else:
        return y
    
reduce(histoCounter, data)                            # reduce와 함수를 사용하여 같은 원소일경우 사전형 value를 1 더함, 아닐경우 다음 원소로 넘어감
input_1=element
print(input_1)
keys=sorted(list(input_1.keys()))
values=list(input_1.values())
for i in range(len(keys)):
    print(keys[i], '\t:','='*values[i])
```
