```python
def prize_list(url):
    year = []
    movie = []
    
    html = urlopen(url)
    soup = BeautifulSoup(html, "lxml")

    tmp = soup.find_all('table')
    tmp1 = tmp[2].find_all('th')
    tmp2 = tmp[2].find_all('td')
    
    for i in range(len(tmp1)):       
        year_tmp = re.split((' '), tmp1[i].get_text())[0]
        year.append(year_tmp)
        if "href" in str(tmp2[i]):
            movie_tmp = tmp2[i].find("a").get_text()
            movie.append(movie_tmp)
        else:
            movie_tmp = tmp2[i].get_text().split()[0]
            movie.append(movie_tmp)
        
    data = {'Year':year,'Movie':movie}
    df = pd.DataFrame(data)
    df.set_index('Year', inplace=True)
    
    return df
    
```

```python
df1 = prize_list('https://movie.naver.com/movie/bi/fi/prize.nhn?code=36&ins=22')
```

```python
def prize_set():
    url = ['https://movie.naver.com/movie/bi/fi/prize.nhn?code=36&ins=22',
          'https://movie.naver.com/movie/bi/fi/prize.nhn?code=3&ins=1',
           
          'https://movie.naver.com/movie/bi/fi/prize.nhn?code=6&ins=1',
           'https://movie.naver.com/movie/bi/fi/prize.nhn?code=6&ins=3',
           'https://movie.naver.com/movie/bi/fi/prize.nhn?code=6&ins=4',
           'https://movie.naver.com/movie/bi/fi/prize.nhn?code=6&ins=5',
           'https://movie.naver.com/movie/bi/fi/prize.nhn?code=6&ins=6',
           'https://movie.naver.com/movie/bi/fi/prize.nhn?code=6&ins=7',
           'https://movie.naver.com/movie/bi/fi/prize.nhn?code=6&ins=11',
           'https://movie.naver.com/movie/bi/fi/prize.nhn?code=6&ins=18',
           'https://movie.naver.com/movie/bi/fi/prize.nhn?code=6&ins=25',
          'https://movie.naver.com/movie/bi/fi/prize.nhn?code=37&ins=2',
          'https://movie.naver.com/movie/bi/fi/prize.nhn?code=33&ins=1',
          'https://movie.naver.com/movie/bi/fi/prize.nhn?code=19&ins=1',
          'https://movie.naver.com/movie/bi/fi/prize.nhn?code=30&ins=1',
          'https://movie.naver.com/movie/bi/fi/prize.nhn?code=18&ins=1']
    prize_name = ['벤쿠버','칸','아카데미-작품','아카데미-남주','아카데미-여주','아카데미-남조','아카데미-여조','아카데미-감독','아카데미-각본','아카데미-음향','아카데미-음악','토론토','베를린','백상','대종상','청룡']
    df=pd.DataFrame()
    for i in range(16):
        df1 = prize_list(url[i])
        df1.rename(columns={'Movie':prize_name[i]}, inplace=True)
        df = pd.concat([df, df1], axis=1, sort=True)
    return df
```

```python
df_prize = prize_set()
```
