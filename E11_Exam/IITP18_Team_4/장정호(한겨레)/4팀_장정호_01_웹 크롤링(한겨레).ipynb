## 한겨레 '4차 산업' 키워드

```python
NAME_4th_Industrial = []
TIME_4th_Industrial = []

url = 'http://search.hani.co.kr/Search?command=query&keyword=4%EC%B0%A8%EC%82%B0%EC%97%85&media=news&sort=d&period=all&datefrom=2018.01.01&dateto=2018.10.31&pageseq={page}'

response1 = urlopen(url.format(page=0))
soup1 = BeautifulSoup(response1,'lxml')
tmp1 = soup1.find_all('dl')       
# len(tmp1)
for lenTmp in range(len(tmp1)):
    if lenTmp == 0:
        tmp3 = tmp1[lenTmp].find('dt').find('a').get_text()
        time3=tmp1[lenTmp].find('dd','date').get_text().strip().split(' ')[0]
    elif lenTmp % 2 == 0:
        tmp3 = tmp1[lenTmp].find('dt').find('a').get_text()
        time3=tmp1[lenTmp].find('dd','date').get_text().strip().split(' ')[0]
    else:
        pass
    TIME_4th_Industrial.append(time3)
    TIME_4th_Industrial1=TIME_4th_Industrial[::2]
    NAME_4th_Industrial.append(tmp3)
    NAME_4th_Industrial1=NAME_4th_Industrial[::2]   

for pg in count():
    page = pg+1
    response = urlopen(url.format(page=page))
    soup = BeautifulSoup(response,'lxml')
    tmp = soup.find_all('dl')
    if page==51:
        break
    
    if len(tmp) is 0:
        print('crawling is finished !!')
        break
        
    
    print('crawling...%d' % page)
    for lenTmp in range(len(tmp)):
        if lenTmp == 0:
            tmp2 = tmp[lenTmp].find('dt').find('a').get_text()
            time2=tmp[lenTmp].find('dd','date').get_text().strip().split(' ')[0]
        elif lenTmp % 2 == 0:
            tmp2 = tmp[lenTmp].find('dt').find('a').get_text()
            time2=tmp[lenTmp].find('dd','date').get_text().strip().split(' ')[0]
        else:
            pass
        TIME_4th_Industrial.append(time2)
        TIME_4th_Industrial1=TIME_4th_Industrial[::2]
        NAME_4th_Industrial.append(tmp2)
        NAME_4th_Industrial1=NAME_4th_Industrial[::2]
```
