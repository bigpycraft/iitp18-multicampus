```python
import re
from konlpy.tag import Hannanum
hannanum = Hannanum()
import nltk
```

```python
final_even['Title']
p = re.compile('\[.*?\]')
for k in range(len(final_even['Title'])):
    real_text = re.sub(p, "", final_even['Title'][k])
    final_even['Title'][k] = real_text

final_even
final_even.set_index('Unnamed: 0')
```

```python
text2 = ". ".join(final_even["Title"])
text2 = text2.replace('세종시','세종')
even_nouns = hannanum.nouns(text2)
```

```python
text1 = ". ".join(final_emp["Title"])
emp_nouns = hannanum.nouns(text1)
```

```python
one_words = []
tow_words = []
etc_words = []
for noun in even_nouns:
    if len(noun) == 1:
        one_words.append(noun)
    elif len(noun) == 2:
        tow_words.append(noun)
    else:
        etc_words.append(noun)
        
len(one_words), len(tow_words), len(etc_words)

stop_words = []
# stop_words = ['제', '월', '일','조','수','때','그','이','바','및','안']
stop_words.extend(one_words)
len(stop_words)

# stop_words 빼기
even_nouns = [word for word in even_nouns if word not in stop_words]
len(even_nouns)

temp  = nltk.Text(even_nouns, name='even')
type(temp), len(temp)

temp.vocab().most_common(20)
```
