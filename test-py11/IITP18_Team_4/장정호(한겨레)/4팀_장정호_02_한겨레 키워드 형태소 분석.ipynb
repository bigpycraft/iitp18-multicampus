```python
p = re.compile('\[.*?\]')
df1 = pd.read_csv('./data/4thIndustry.csv', encoding='UTF-8')
def 핸들링(data):
    
    Title1 = list(data['Title'])

    for i in range(len(Title1)):
        real_text = re.sub(p,'',Title1[i])
        Title1[i] = real_text
# print(Title1)

    text = ",".join(Title1)
    text = text.replace('…','').replace('/','').replace('-','').replace('·',',').replace('≠','').replace('(','').replace(')','').replace('?','').replace('!','').replace('~','').replace('㈜','').replace('  ','').replace('...','').replace("‘",'').replace("’",'').replace('“','').replace('”','')
    return text
```
