# CNN

```python
# 소요 시간 측정 함수
import time
import os

def chk_pocessing_time(start_time, end_time):
    process_time = end_time - start_time
    p_time = int(process_time)
    p_min = p_time // 60
    p_sec = p_time % 60
    print('처리시간 : {p_min}분 {p_sec}초 경과되었습니다.'.format(
    p_min = p_min,
    p_sec = p_sec))
    return process_time
```

```python
import tensorflow as tf
import random
import matplotlib.pyplot as plt
tf.set_random_seed(777)
```

## 1. Convolution Layer1
* learning_rate = 0.001
* training_epochs = 500
* batch_size = 100
* <b>input:</b> 96 * 96 image, <b>output:</b> 30 keypoints

```python
# hyper parameters 정의
learning_rate = 0.001
training_epochs = 500
batch_size = 100

# input placeholders... x값은 96*96 이미지픽셀값, y값은 30개의 예측값!
X = tf.placeholder(tf.float32, [None, 9216])# 96*96 = 9216
X_img = tf.reshape(X, [-1, 96, 96, 1]) # 텐서플로우가 인식할 수 있도록 4차원으로 rechape
Y = tf.placeholder(tf.float32, [None, 30])

# L1 ImgIn shape = (?, 96, 96, 1)
# 먼저 마스크 사이즈를 정의해주자 --> 3*3 사이즈 64개!
W1 = tf.Variable(tf.random_normal([3, 3, 1, 64], stddev= 0.01))
print('W1 (tf.random_nomal) \t: ', W1)
# Conv --> (?, 96, 96, 64)
# Pool --> (?, ...)

# convolution 정의 ... 1칸씩 움직인다!
L1 = tf.nn.conv2d(X_img, W1, strides=[1, 1, 1, 1], padding='SAME')
print('tf.nn,conv2d \t:', L1)

# activation 함수는 relu를 적용
L1 = tf.nn.relu(L1)
print('tf.nn.relu \t:', L1)

# pooling 
L1 = tf.nn.max_pool(L1, ksize = [1, 2, 2, 1], strides=[1, 2, 2, 1], padding = 'SAME')
print('tf.nn.max_pool \t: ', L1)
```

## 2. Convolution Layer2

```python
# L2 ImgIn shape = (?, 48, 48, 16)
W2 = tf.Variable(tf.random_normal([3, 3, 64, 128], stddev = 0.01))
print('W2 (tf.random_normal) \t: ', W2)
# Conv --> (?, 48, 48, 64)
# Pool --> (?, 24, 24, 64)
L2 = tf.nn.conv2d(L1, W2, strides=[1, 1, 1, 1], padding = 'SAME')
print('tf.nn.conv2d \t:', L2)
L2 = tf.nn.relu(L2)
print('tf.nn.relu \t:', L2)
L2 = tf.nn.max_pool(L2, ksize = [1, 2, 2, 1], strides=[1,2,2,1], padding = 'SAME')
print('tf.nn.max_pool \t:', L2)

# 1차원으로 Flattening
L2_flat = tf.reshape(L2, [-1, 24 * 24 * 128]) 
print('tf.reshape \t:', L2_flat)

'''
Tensor("Conv2D_1:0", shape=(?, 48, 48, 64), dtype=float32)
Tensor("Relu_1:0", shape=(?, 48, 48, 64), dtype=float32)
Tensor("MaxPool_1:0", shape=(?, 24, 24, 64), dtype=float32)
Tensor("Reshape_1:0", shape=(?, 36864), dtype=float32)
'''
```

## 3. Fully Connected(FC, Dense) Layer
input: 최종적으로 Fully connected Layer는 24\*24\*128 형태 --> output: 30개의 예측값

```python
W11 = tf.get_variable("W11", shape = [24*24 * 128, 30], initializer=tf.contrib.layers.xavier_initializer())
print('W11 (xavier_intializer) \t:', W11)

b = tf.Variable(tf.random_normal([30]))
logits = tf.matmul(L2_flat, W11) + b
```

### 1) cost & optimizer 정의
>- cost
>    - 1차: nn.softmax_cross_entropy_with_logits(...) 
>    - 2차: MSE 수식으로 변경 (tf.square(logits - Y)) --> <font color = 'Blue'>cost 대폭 감소</font>

>- optimizer
>    - AdamOptimizer

```python
cost = tf.reduce_mean(tf.square(logits - Y))
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)

print( 'cost :', cost)
print('optimizer :\n', optimizer)
```

### 2) Model training

```python
# 초기화
sess = tf.Session()
sess.run(tf.global_variables_initializer())

# train my model
print('Learning started. It takes sometime.')
time1 = time.time()

for epoch in range(training_epochs):
    avg_cost = 0
    total_batch = int( len(data)/ batch_size)
    
    for i in range(total_batch):
        batch_xs = data.Image[i*batch_size: (i+1)*batch_size]
        batch_xs = np.vstack([np.expand_dims(x, 0) for x in batch_xs]) ## 데이터 형태 변경
        batch_ys = data.iloc[:, :-1][i*batch_size: (i+1)*batch_size] # array 타입으로 가져옴
        feed_dict = {X: batch_xs, Y: batch_ys}
        c, _ = sess.run([cost, optimizer], feed_dict = feed_dict)
        avg_cost += c / total_batch
    print('Epoch:', '%04d' % (epoch +1), 'cost =', '{:.9f}'.format(avg_cost))
print('Learning Finished!')
time2 = time.time()

chk_pocessing_time(time1, time2)
```
